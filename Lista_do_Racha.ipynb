## Função para construir a lista

```python
# data_sheet.iloc[4,1] == "Veterano"
def build_list(df):
  list_main = []
  list_suplentes = []
  N = len(data_sheet) # <------ Get the list length automatically
  N_max = 30
  for ii in range(N):
    
    if ii < N_max:
      if data_sheet.iloc[ii,1] == "Veterano":
        list_main.append(data_sheet.iloc[ii,2])    
        if data_sheet.iloc[ii,3] == "Sim":
          if data_sheet.iloc[ii,4] == "Veterano":
            list_main.append(data_sheet.iloc[ii,5])
          else:
            list_suplentes.append(data_sheet.iloc[ii,5] + " ({})".format(data_sheet.iloc[ii,2])) # <------ Save the name associated (veterano)
      if data_sheet.iloc[ii,1] == "Convidado":
        list_suplentes.append(data_sheet.iloc[ii,2]) # <------ Save the name associated (veterano)
    else:
      list_suplentes.append(data_sheet.iloc[ii,2])

  return(list_main, list_suplentes)

def print_list(principal, suplentes, date):
  print("*Lista R9*\n{}\n{} bola ao centro\n\n*Goleiros*:\n1 - \n2 - \n\n*Jogadores*:".format(date, "*20h30*"))

  N_principal = len(principal);  

  
  '''
  Para imprimir 18, 24, ou 30 espaços usar o seguinte bloco.
  Comentar este bloco caso queira imprimir todos os 30 espaços
  '''
  if N_principal < 18: 
    N_max = 18
  elif N_principal > 18 and N_principal < 24:
    N_max = 24
  else:
    N_max = 30
  
  '''
  Para imprimir todos os 30 espaços basta descomentar a linha seguinte
  '''
  # N_max = 30

  # Formatacao da lista, posicao de inscricao na lista e nome do membro
  if N_principal != N_max: 
    for index in range(N_principal):
      print('{} - {}'.format(index+1, principal[index] ))
    for ii in range(N_max - N_principal):
      print('{} - '.format(N_principal+ii+1))
  
  # Lista de suplentes
  print("\n*Suplentes*:") 
  for index in range(len(suplentes)):
    print('{}'.format(suplentes[index]))
```

# Gerar a lista

```python
DATA_DO_RACHA = 'Quarta {:%d/%m}'.format(proxima_quarta) # <-------------------- DATA AUTOMATICA
print_list(build_list(data_sheet)[0], build_list(data_sheet)[1], date=DATA_DO_RACHA)
```
